```python
#1. Theoritical Derivation(Analytical Method)
#2. Simulation Method (Empirical Method) to calculate the sample distribution

import numpy as np
import matplotlib.pyplot as plt

# population parameters
mu = 160
sigma = 10
n = 25
num_samples = 10000

# simulate 10000 samples of size 25
sample_means = [np.mean(np.random.normal(mu, sigma, n)) for _ in range(num_samples)]
print(sample_means)

# plot histogram
plt.hist(sample_means, bins=50, density=True, color='skyblue', edgecolor='black')
plt.title('Sampling Distribution of Sample Mean (Simulated)')
plt.xlabel('Sample Mean Height (cm)')
plt.ylabel('Density')
plt.show()
```

```python
# Bootstrap method (FResampling from sample)

# take a real sample
sample = np.random.normal(160, 10, 25)

# bootstrap: resample with replacement 10000 times
bootstrap_means = [np.mean(np.random.choice(sample, size=25, replace=True)) for _ in range(10000)]

# plot histogram
plt.hist(bootstrap_means, bins=50, density=True, color='lightgreen', edgecolor='black')
plt.title('Bootstrap Sampling Distribution of Mean')
plt.xlabel('Bootstrapped Sample Mean')
plt.ylabel('Density')
plt.show()
```
